<h2>3.1.3. Lists</h2>

<p>Python knows a number of compound data types, used to group together other values. The most versatile is the list, which can be written as a list of comma-separated values (items) between square brackets. Lists might contain items of different types, but usually the items all have the same type.</p>

```python
squares = [1, 4, 9, 16, 25]
```

```python
squares
```

<p>Like strings (and all other built-in sequence types), lists can be indexed and sliced:</p>

```python
squares[0]  # indexing returns the item
```

```python
squares[-1]
```

```python
squares[-3:]  # slicing returns a new list
```

```python
squares + [36, 49, 64, 81, 100]  #Lists also support operations like concatenation:
```

<p>Unlike strings, which are immutable, lists are a mutable type, i.e. it is possible to change their content:</p>

```python
cubes = [1, 8, 27, 65, 125]
```

```python
4 ** 3  # the cube of 4 is 64, not 65!
```

```python
cubes[3] = 64  # replace the wrong value
```

```python
cubes
```

<p>You can also add new items at the end of the list, by using the list.append()</p>

```python
cubes.append(216)  # add the cube of 6
```

```python
cubes.append(7 ** 3)  # and the cube of 7
```

```python
cubes
```

<p>Simple assignment in Python never copies data. When you assign a list to a variable, the variable refers to the existing list. Any changes you make to the list through one variable will be seen through all other variables that refer to it.</p>

```python
rgb = ["Red", "Green", "Blue"]
```

```python
rgba = rgb
```

```python
id(rgb) == id(rgba)  # they reference the same object
```

```python
rgba.append("Alph")
```

```python
rgb
```

<p>All slice operations return a new list containing the requested elements. This means that the following slice returns a shallow copy of the list:</p>

```python
correct_rgba = rgba[:]
```

```python
correct_rgba[-1] = "Alpha"
```

```python
correct_rgba
```

```python
rgba
```

<p>Assignment to slices is also possible, and this can even change the size of the list or clear it entirely:</p>

```python
letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
```

```python
letters
```

```python
# replace some values
letters[2:5] = ['C', 'D', 'E']
```

```python
letters
```

```python
# now remove them
letters[2:5] = []
letters
```

```python
letters[:] = []
letters
```

<p>The built-in function len() also applies to lists:</p>

```python
letters = ['a', 'b', 'c', 'd']
len(letters)
```

<p>It is possible to nest lists (create lists containing other lists), for example:</p>

```python
a = ['a', 'b', 'c']
n = [1, 2, 3]
x = [a, n]
x
```

```python
x[0]
```

```python
x[0][1]
```
